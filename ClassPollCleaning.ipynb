```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/jingchengwu/Desktop/LBS/Data Viz/GroupProject/ClassPollResults.csv")
```

```python
answers = df.iloc[:, 17:38]
answers.head()
```

```python
for i in range(7):
    old_column_name = answers.columns[i]
    answers = answers.rename(columns={old_column_name: answers.iloc[0,i].split('-')[1]})
```

```python
answers = answers.rename(columns={'Q3_0_GROUP':'Top5'})
answers.head()
```

```python
rename_mapping = {}
for i in range(8, answers.shape[1] - 1):
    old_column_name = answers.columns[i]
    new_column_name = "-".join(answers.iloc[0, i].split('-')[-2:]).strip()
    rename_mapping[old_column_name] = new_column_name

answers = answers.rename(columns=rename_mapping)
answers.head()
```

```python
answers = answers.rename(columns={'Q4':'Nomination'})
answers.head()
```

```python
answers = answers.iloc[2:]
```

```python
answers = answers.reset_index()
answers.head()
```

```python
answers.to_csv('Cleaned_PollResults.csv')
```
